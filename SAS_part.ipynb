```python
inter_df=pd.read_csv("interstate.csv")
```

```python
inter_df
```

```python
df_SAS=inter_df.iloc[:,[6,7,8,9,10,11,12,13]]
```

```python
df_SAS
```

```python
df_SAS.NAICS=inter_df.NAICS.astype(str)
df_SAS.QUARTER=inter_df.QUARTER.astype(str)
df_SAS.MODE=inter_df.MODE.astype(str)

df_SAS=df_SAS[df_SAS['MODE']!="2"]

df_SAS=df_SAS[df_SAS['MODE']!="12"]
```

```python
df_SAS['TEMP_CNTR_YN']=df_SAS.apply(lambda r:1 if r.TEMP_CNTL_YN=="Y" else 0, axis=1) 
df_SAS['HAZMAT']=df_SAS.apply(lambda r:1 if (r.HAZMAT=="P" or r.HAZMAT=="H") else 0, axis=1)
```

```python

df_SAS['MODE']=df_SAS.apply(lambda r:1 if (r.MODE=="3" or r.MODE=="4" or r.MODE=="5") else 2 if (r.MODE=="6")
                             else 4 if (r.MODE=="11") else 5 if (r.MODE=="13" or r.MODE=="14" or  r.MODE=="15" or r.MODE=="16" or r.MODE=="17" or
                                                  r.MODE=="18" or r.MODE=="20") else 3 if (r.MODE=="10") else "",axis=1)

df_SAS['NAICS']=df_SAS.apply(lambda r:1 if r.NAICS=="212" else 2 if (r.NAICS[0]=="3" and r.NAICS[1]=="1") or
                                         (r.NAICS[0]=="3" and r.NAICS[1]=="2") or
                                         (r.NAICS[0]=="3" and r.NAICS[1]=="3")
                             else 3 if (r.NAICS[0]=="4" and r.NAICS[1]=="2")
                             else 4 if (r.NAICS[0]=="4" and r.NAICS[1]=="5")
                             else 5 if (r.NAICS[0]=="4" and r.NAICS[1]=="9")
                             else 6 if (r.NAICS[0]=="5" and r.NAICS[1]=="1")
                             else 7 if (r.NAICS[0]=="5" and r.NAICS[1]=="5")
                             else "",axis=1)
```

```python
#splitting into train and validation for parameter selection
from sklearn.model_selection import train_test_split
df_SAS_train, df_SAS_test = train_test_split(df_SAS, test_size = 0.3, random_state = 0)
```

```python
df_SAS.drop(columns="TEMP_CNTL_YN")
```

```python
df_SAS_train.drop(columns="TEMP_CNTL_YN",inplace=True)
df_SAS_test.drop(columns="TEMP_CNTL_YN",inplace=True)
```

```python
df_SAS_train.columns
```

```python
df_SAS_train.to_csv("df_SAS_train.csv")
```

```python
df_SAS_test.to_csv("df_SAS_test.csv")
```
